```python
import pandas as pd
```

```python
pd.DataFrame()
```

```python
# create dataframe from scratch 

raw_data = {
    'name':[ 'nut','nine','gift','grape','suguxs'],
    'age' : [32,22,22,21,23],
    'gender' : ['M','F','F','F','F']
}

df = pd.DataFrame(raw_data)
df
```

```python
# create new columns
df['city'] = ['Bangkok','Bangkok','UK','Bangkok','UK']
```

```python
df
```

```python
## drop columns .drop
df = df.drop('city', axis = 1)
```

```python
df
```

```python
## drop low
df = df.drop(4,axis = 0)
```

```python
df
```

```python
## set index กรณีที่ลบ row ข้างในและrow number ไม่เปลี่ยน
df = df.reset_index(drop = True )
df
```

```python
df
```

```python
# columns renames 
list(df.columns)
```

```python
df.columns = ['nickname','age','sex','city']
df
```

```python
## create a new series 
s1 = pd.Series(['mary',20,'F','USA'],index =['nickname','age','sex','city'])
print(s1)
print(type(s1))
```

```python
#append series to dataframe
df.append(s1,ignore_index = True)
```

```python
## apppend row data 
s2 = pd.Series([3.55,4.0,3.45,3.8,3.5])
df['gpa'] = s2 
df
```

```python
# write csv file 
df.to_csv('mydata.csv)
```
